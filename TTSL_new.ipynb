```python
def highlight_text_first_cin_rows(df):
    mask = ~df.duplicated(subset='Company ID', keep='first')
    
    return pd.DataFrame(
        [['color: #0070C0' if mask.iloc[i] else '' for _ in df.columns]
         for i in range(len(df))],
        index=df.index,
        columns=df.columns
    )
```

```python
styled = final.style.apply(highlight_text_first_cin_rows, axis=None)
styled
```
